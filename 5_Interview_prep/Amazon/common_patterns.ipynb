```python
# Hashmaps (counting problem)

from collections import defaultdict

def count_elements(nums):
    count = defaultdict(int)
    for x in nums:
        count[x]+=1
    for k,v in count.items():
        print(k,v)
```

```python
# Two Pointers (sorted array)

def two_sum(nums, target):
    l, r = 0, len(nums)-1
    while l < r:
        s = nums[l]+ nums[r]
        if s == target:
            return[l, r]
        elif s < target:
            l += 1
        else:
            r -= 1
```
