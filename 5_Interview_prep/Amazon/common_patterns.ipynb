```python
# Hashmaps (counting problem)

from collections import defaultdict

def count_elements(nums):
    count = defaultdict(int)
    for x in nums:
        count[x]+=1
    for k,v in count.items():
        print(k,v)
```

```python
# Two Pointers (sorted array)

def two_sum(nums, target):
    l, r = 0, len(nums)-1
    while l < r:
        s = nums[l]+ nums[r]
        if s == target:
            return[l, r]
        elif s < target:
            l += 1
        else:
            r -= 1
```

```python
# Sliding Window Pattern

l = 0
for r in range(len(nums)):
    # do something with nums[r]
    while condition:
        # do something with nums[l]
        l += 1  
```

```python
# Binary Search

def binary_search(nums, target):
    l, r = 0, len(nums)-1
    while l <= r:
        m = l + (r-l)//2
        if nums[m]== target:
            return m
        elif nums[m] < target:
            l = m + 1
        else:
            r = m - 1
```

```python
# DFS (tree)

def dfs(node):
    if not node:
        return
    # do something with node
    dfs(node.left)
    dfs(node.right)
```

```python
# BFS (tree)

from collections import deque

q = deque
while q:
    node = q.popleft()
    # do something with node
    if node.left:
        q.append(node.left)
    if node.right:
        q.append(node.right)
```

```python
# DFS (graph)

def dfs(node, visited):
    if node in visited:
        return
    visited.add(node)
    # do something with node
    for neighbor in node.neighbors:
        dfs(neighbor, visited)
```
