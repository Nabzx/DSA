# Binary Search

### Definition
- Works on sorted data (or monotonic condition)
- Repeatedly halves search space

### Time Complexity
O(log n)

```python
# Traditional Binary Search
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1
```

### Idea
We are NOT searching for a value

We are searching for the FIRST index where condition = TRUE

```python
# Template for Binary Search with a Condition
def binary_search_condition(low, high):
    while low < high:
        mid = (low + high) // 2
        
        if condition(mid):
            high = mid
        else:
            low = mid + 1
    
    return low
```

### When to use
- Sorted array → traditional
- Answer space → condition-based

Examples:
- Minimum capacity
- First bad version
- Allocate books
- Minimize max value

### Common mistakes
- Using <= instead of <
- Infinite loops
- Wrong mid update
- Off-by-one errors
