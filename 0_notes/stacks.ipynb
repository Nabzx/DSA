# Stacks

### Definition
- LIFO data structure
- Last element added is first removed

### Operations
- push → add
- pop → remove
- peek → top element

```python
# Python Usage


# Create stack
stack = []

# Push
stack.append(1)
stack.append(2)

# Pop
stack.pop()  # removes 2

# Peek
stack[-1]

# Check empty
if not stack:
    print("empty")
```

### Time Complexity
Push → O(1)  
Pop → O(1)  
Peek → O(1)

### Common Use Cases
- Matching parentheses
- Undo operations
- Expression evaluation
- Monotonic stack problems

```python
# Basic Template

def use_stack(nums):
    stack = []
    
    for num in nums:
        stack.append(num)
    
    while stack:
        print(stack.pop())
```

### Key Insights

Stacks are used when:
- You need to process things in reverse
- You care about the most recent element
