# Linked Lists

### Definition
- Collection of nodes
- Each node contains value + pointer(s)

### Types
- Singly: next pointer
- Doubly: prev + next

```python
# Node Structure
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# Doubly
class DoublyNode:
    def __init__(self, val=0):
        self.val = val
        self.prev = None
        self.next = None
```

### Time Complexity
Access → O(n)  
Insert (head) → O(1)  
Delete (head) → O(1)  
Search → O(n)

### Key Concepts
- Use pointers, not indices
- Always track prev and next
- Be careful with None (end of list)

- Empty list
