# Queues

### Definition
- FIFO data structure
- First element added is first removed

### Operations
- enqueue → add
- dequeue → remove

```python
# Python Usage


from collections import deque

q = deque()

# enqueue
q.append(1)
q.append(2)

# dequeue
q.popleft()  # removes 1

# peek
q[0]

# check empty
if not q:
    print("empty")
```

### Time Complexity

Enqueue → O(1)  
Dequeue → O(1)  
Peek → O(1)

### Common Use Cases

- Breadth First Search (BFS)
- Level order traversal
- Scheduling
- Sliding window

### Key Insight

Queues are used when:
- You process things in order
- You explore level by level
