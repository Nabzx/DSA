# Arrays

### Definition
- Collection of elements stored in contiguous memory
- Access using index

### Properties
- Ordered + Contiguous
- Index-based
- Mutable -> they can be changeable (e.g. size, number of items etc.)

### Time Complexity
Access → O(1)  
Search → O(n)  
Insert (end) → O(1)  
Insert (middle) → O(n)  
Delete → O(n)

```python
# Basic Operations


arr = [1, 2, 3]

# Access
print(arr[0])

# Append
arr.append(4)

# Remove
arr.pop()

# Iterate
for x in arr:
    print(x)
```

### Common Patterns
- Sliding window
- Two pointers
- Prefix sum
- Binary search (on sorted arrays)

### Key Tricks
- Use index instead of modifying array
- Avoid nested loops if possible
- Think in terms of pointers

### Edge Cases
- Empty array
- Single element
- All same elements
- Negative numbers
