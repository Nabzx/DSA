# Graphs

### Definition
- Nodes (vertices) connected by edges

### Types
- Directed / Undirected
- Weighted / Unweighted
- Cyclic / Acyclic

```python
# Representation of a graph using an adjacency list
graph = {
    0: [1, 2],
    1: [0, 3],
    2: [0],
    3: [1]
}
```

```python
# Using defaultdict
from collections import defaultdict

graph = defaultdict(list)

graph[0].append(1)
graph[1].append(0)
```

### Traversals

DFS → depth-first  
BFS → level-by-level  

### Key concepts

- Visited set (avoid cycles)
- Connected components
- Cycles
- Paths

### When to use BFS vs DFS

BFS:
- shortest path (unweighted)
- level traversal

DFS:
- explore all paths
- recursion problems

### Key Insight

Graph problems = traversal + visited tracking
