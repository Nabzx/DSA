# Heaps

## Definition

- Complete binary tree
- 2 types which are minHeap and maxHeap
- minHeap: parent < child
- maxHeap: parent > child
- Right child 2i + 1
- Left child 2i + 2




```python
import heapq

# Min heap
heap = []
heapq.heappush(heap, 3)
heapq.heappush(heap, 1)
heapq.heappop(heap) 

# Max heap (hack)
heapq.heappush(heap, -5)
```

## Time complexities
Push → O(log n)  
Pop → O(log n)  
Peek → O(1)  
Heapify → o(n)  
Heap sort → o(nlogn)  

## Most common usages
- Top K problems
- Priority queue
- Merge K sorted lists
- Dijkstra’s algorithm

```python
# Basic template
def use_heap(nums):
    heap = []
    
    for num in nums:
        heapq.heappush(heap, num)
    
    while heap:
        print(heapq.heappop(heap))
```
