# Hash Tables (Maps & Sets)

### Definition
- Store key → value pairs
- Use hashing for O(1) average access

### Types
- Map (dict): key → value
- Set: unique elements only

### What is hashable? 
- Yes: Strings, Integers, Tuples (They are immutable)
- No: Arrays, Dictionaries
- This is to ensure consistency in the table. Mutable objects can be changed.

### Collisions
- Separate chaining (linked lists inside table)
- Linear probing (goes to next index position if full)

```python
# Map
d = {}
d["a"] = 1
d["b"] = 2

print(d["a"])      # 1
print(d.get("c"))  # None (safe)

# Set
s = set()
s.add(1)
s.add(2)
s.add(1)  # ignored

print(1 in s)  # True
```

### Common Operators

Insert → O(1)  
Delete → O(1)  
Lookup → O(1)

### Patterns/Use cases

- Counting frequencies  
- Checking duplicates   
- Fast lookup   
- Complement search (Two Sum)

```python
# Frequency map
def count_freq(nums):
    freq = {}
    
    for num in nums:
        freq[num] = freq.get(num, 0) + 1
    
    return freq
```

```python
# Set for up
def contains_duplicate(nums):
    seen = set()
    
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    
    return False
```
