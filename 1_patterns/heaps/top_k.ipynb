# Top K Pattern

### When to use
- Find K largest/smallest elements

### Idea
- Use min heap of size K

```python
# Template
import heapq

def top_k(nums, k):
    heap = []
    
    for num in nums:
        heapq.heappush(heap, num)
        
        if len(heap) > k:
            heapq.heappop(heap)
    
    return heap
```

```python
# Practice

# Problem: Kth Largest Element in Array

import heapq

def findKthLargest(nums, k):
    heap = []
    
    for num in nums:
        heapq.heappush(heap, num)
        if len(heap) > k:
            heapq.heappop(heap)
    
    return heap[0]
```
