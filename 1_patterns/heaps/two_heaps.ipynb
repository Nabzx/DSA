# Two Heaps

### Used for - Median Problems

```python
# Template

import heapq

small = []  # max heap (store negatives)
large = []  # min heap

def add_num(num):
    heapq.heappush(small, -num)
    
    heapq.heappush(large, -heapq.heappop(small))
    
    if len(large) > len(small):
        heapq.heappush(small, -heapq.heappop(large))
```
