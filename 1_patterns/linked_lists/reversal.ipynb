# Reverse Linked List

### Idea
Reverse pointers one by one

```python
# Template

def reverse(head):
    prev = None
    curr = head
    
    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt
    
    return prev
```
