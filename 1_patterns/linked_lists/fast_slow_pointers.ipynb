# Fast & Slow Pointers

### When to use
- Find middle
- Detect cycles

### Idea
- slow moves 1 step
- fast moves 2 steps

```python
# Template

def middle(head):
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    
    return slow
```

```python
# Practice

# Find middle of linked list

def middleNode(head):
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    
    return slow
```
