# Two Sum Pattern

### When to use
- Find pair with target
- Complement lookup

### Idea
Store seen numbers in hashmap

```python
# Template

def two_sum(nums, target):
    seen = {}
    
    for i, num in enumerate(nums):
        diff = target - num
        
        if diff in seen:
            return [seen[diff], i]
        
        seen[num] = i
```

```python
# Practice

# Problem: Two Sum

def twoSum(nums, target):
    seen = {}
    
    for i, num in enumerate(nums):
        if target - num in seen:
            return [seen[target - num], i]
        seen[num] = i
```
