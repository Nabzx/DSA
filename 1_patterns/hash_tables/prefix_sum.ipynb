### Idea
- Use hashmap to store prefix sums

```python
# Template
def subarray_sum(nums, k):
    prefix = 0
    count = 0
    seen = {0: 1}
    
    for num in nums:
        prefix += num
        
        if prefix - k in seen:
            count += seen[prefix - k]
        
        seen[prefix] = seen.get(prefix, 0) + 1
    
    return count
```
