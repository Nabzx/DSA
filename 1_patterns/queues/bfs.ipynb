# Breadth First Search (BFS)

### When to use
- Trees / graphs
- Shortest path (unweighted)
- Level-order traversal

### Idea
Process nodes level by level using a queue

```python
# Template

from collections import deque

def bfs(start):
    q = deque([start])
    visited = set([start])
    
    while q:
        node = q.popleft()
        
        for neighbor in node.neighbors:
            if neighbor not in visited:
                visited.add(neighbor)
                q.append(neighbor)
```

```python
# Practice

# Binary Tree Level Order Traversal

from collections import deque

def levelOrder(root):
    if not root:
        return []
    
    q = deque([root])
    result = []
    
    while q:
        level = []
        
        for _ in range(len(q)):
            node = q.popleft()
            level.append(node.val)
            
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
        
        result.append(level)
    
    return result
```
