# Sliding Window (Queue)

### When to use
- Fixed-size window
- Process elements in order

### Idea
Add new → remove old

```python
# Template

from collections import deque

def window(nums, k):
    q = deque()
    
    for i in range(len(nums)):
        q.append(nums[i])
        
        if len(q) > k:
            q.popleft()
```

```python
# Practice

# Moving Average

from collections import deque

class MovingAverage:
    def __init__(self, size):
        self.q = deque()
        self.size = size
        self.total = 0
    
    def next(self, val):
        self.q.append(val)
        self.total += val
        
        if len(self.q) > self.size:
            self.total -= self.q.popleft()
        
        return self.total / len(self.q)
```
