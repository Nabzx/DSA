- Contiguous elements

### Idea
Expand + shrink window

```python
# Practice

# Problem: Maximum Subarray Sum of Size K

def maxSum(nums, k):
    window_sum = 0
    max_sum = 0
    left = 0
    
    for right in range(len(nums)):
        window_sum += nums[right]
        
        if right - left + 1 == k:
            max_sum = max(max_sum, window_sum)
            window_sum -= nums[left]
            left += 1
    
    return max_sum
```
