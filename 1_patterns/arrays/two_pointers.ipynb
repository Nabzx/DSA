# Two Pointers

### When to use
- Sorted arrays
- Pairs / triplets
- Opposite ends

### Idea
Use left + right pointer

```python
# Template

def two_pointers(arr):
    left, right = 0, len(arr) - 1
    
    while left < right:
        if condition:
            left += 1
        else:
            right -= 1
```

```python
# Practice

# Problem: Two Sum II (sorted array)

def twoSum(numbers, target):
    left, right = 0, len(numbers) - 1
    
    while left < right:
        s = numbers[left] + numbers[right]
        
        if s == target:
            return [left + 1, right + 1]
        elif s < target:
            left += 1
        else:
            right -= 1
```
