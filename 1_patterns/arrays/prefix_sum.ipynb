# Prefix Sum

### Idea
Store cumulative sums to answer queries fast

```python
# Template

def prefix_sum(nums):
    prefix = [0] * len(nums)
    prefix[0] = nums[0]
    
    for i in range(1, len(nums)):
        prefix[i] = prefix[i-1] + nums[i]
    
    return prefix
```

```python
# Practice

# Problem: Range Sum Query

def rangeSum(nums, left, right):
    prefix = [0]
    
    for num in nums:
        prefix.append(prefix[-1] + num)
    
    return prefix[right + 1] - prefix[left]
```
